```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Seal AM de entrada
t = np.linspace(0, 1, 44100)  # Tiempo
fc = 1000  # Frecuencia de la portadora
fm = 10  # Frecuencia de la seal moduladora
modulation_index = 0.8  # ndice de modulacin
carrier = 10*np.sin(2 * np.pi * fc * t)  # Seal portadora
modulating_signal = np.sin(2 * np.pi * fm * t)  # Seal moduladora
modulated_signal = (1 + modulation_index * modulating_signal) * carrier  # Seal modulada

# Demodulacin AM
envelope = np.abs(modulated_signal)  # Envolvente
lpf_cutoff = 200  # Frecuencia de corte del filtro pasa bajos
lpf_order = 3  # Orden del filtro pasa bajos
b, a = signal.butter(lpf_order, lpf_cutoff / (44100 / 2), 'lowpass')  # Coeficientes del filtro pasa bajos
demodulated_signal = signal.filtfilt(b, a, envelope)  # Seal demodulada

# Graficar la seal demodulada
fig, ax = plt.subplots()
ax.plot(t, demodulated_signal, label='Seal demodulada')
ax.set_xlabel('Tiempo (s)')
ax.set_ylabel('Amplitud')
ax.legend()
plt.show()
# Graficar la seal modulada
fig, ax = plt.subplots()
ax.plot(t, modulated_signal, label='Seal modulada')
ax.set_xlabel('Tiempo (s)')
ax.set_ylabel('Amplitud')
ax.legend()
plt.show()

```
